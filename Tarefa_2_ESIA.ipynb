## 1. Importe a biblioteca numpy, pandas, bokeh io e plotting.

```
## Implemente o codigo iniciando na linha abaixo ( Uma linha de codigo esperada)
import pandas as pd
import numpy as np
from bokeh.io import output_notebook
from bokeh.plotting import figure, show
```

## 2. Crie um Data Frame df a partir do csv 'clima.csv'

```
## Implemente o codigo iniciando na linha abaixo ( Uma linha de codigo esperada)

output_notebook()

df = df = pd.read_csv('/content/clima-atividade 2.csv')

df.tail(9)
```

## 3. Agrupe por ano as médias das temperaturas médias do dataframe e atribua a um novo data frame registros_por_ano que possua só as colunas com informações das temperaturas média e o ano.

```

registros_por_ano = pd.DataFrame({'count' : df.groupby(["tmedia","ano"]).size()}).reset_index()

registros_por_ano.columns = ["tmedia","ano","total"]

df.groupby('ano').count()


registros_por_ano.head(50)
```
